```python
cat = data.chelsea()
print('Shape:', cat.shape)
print('Values min/max:', cat.min(), cat.max())
plt.imshow(cat);
```

```python
from skimage import io, segmentation as seg, color
```

```python
labels = seg.slic(cat, n_segments=8, compactness=5)
plt.imshow(labels)

print(labels)
```

```python
def mean_color(cat, labels):
    out = np.zeros_like(cat)
    for label in np.unique(labels):
        indices = np.nonzero(labels == label)
    out[indices] = np.mean(cat[indices], axis = 0)
    return out
```

```python
labels = seg.slic(cat, n_segments=8, compactness=5, enforce_connectivity=True)
label_image = mean_color(cat, labels)
plt.imshow(label_image)

```

```python
labels = seg.slic(cat, n_segments=18, compactness=5, enforce_connectivity=True)
label_image = mean_color(cat, labels)
plt.imshow(cat)
```
